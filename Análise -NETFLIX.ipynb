# Análise exploratória - Netflix

```python
#importando bibliotecas
import pandas as pd
import plotly.express as px
```

```python
#Ler o arquivo
tabela = pd.read_csv("netflix_titles.csv")
```

```python
#Visualizar arquivo
display(tabela)
```

```python
#Visualizar informações
print(tabela.info())
```

```python
#Descartar informações que atrapalham
tabela = tabela.drop("description", axis = 1)
tabela = tabela.drop("show_id", axis = 1)
tabela = tabela.drop("cast", axis = 1)
```

```python
#Verificando novamente se existe a necesidade de transformar mais dados
display(tabela)
```

```python
# GRÁFICOS
grafico = px.histogram(tabela, x= "type", text_auto = True)
grafico.show()
```

```python
grafico1 = px.histogram(tabela, x= "release_year", text_auto = True )
grafico1.show()
```

```python
grafico3 = px.histogram(tabela, x = 'country')
grafico3.show()
```

```python
grafico4 = px.histogram(tabela, x = 'rating')
grafico4.show()
#"TV-MA", o programa não é adequado para menores de 17 anos.
#TV-14. Os pais estão dispostos a exercer um maior cuidado no acompanhamento do programa e são advertidos contra deixar as crianças de qualquer idade nas horas muito tarde.
```
